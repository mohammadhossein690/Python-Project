```python
def are_you_ready():
    '''
    This is are you ready function
    '''
    while True:
        r = input('Are you ready (Yes-No): ')
        if r in ['Yes','No']:
            break
        else:
            print('Please Enter Yes or No')
    return r
def welcome_back():
    '''
    This is welcome back function
    '''
    while True:
        r = input('Do you wanna play again? (Yes-No): ')
        if r in ['Yes','No']:
            break
        else:
            print('Please Enter Yes or No')
    return r
def display(row1,row2,row3):
    '''
    This is display function
    '''
    print('This is your current board : ')
    print()    
    print(row1)
    print(row2)
    print(row3)
def position_check():
    '''
    This is position check function
    '''
    choice = ''
    within_range = False
    while choice.isdigit() == False or within_range == False:
        choice = input('Choose position from 1-9 : ')
        if choice.isdigit() == False:
            print('Please Enter digit!!!')
        else:
            if int(choice) in range(1,10):
                within_range = True
            else:
                print('Out of range!!!')
    return int(choice)
```
