```python
# First we import our needed modules
import string
import random
import time
import winsound
```

```python
def user_check() -> int:
    '''
    This function will get a number
    from user and create password 
    base on this number
    '''
    while True:
        try:
            num = int(input('Enter lenght of password : '))
        except:
            print('Please Enter number')
        else:
            break
    return num
```

```python
#Create password
AVAILABLE  = string.ascii_letters + string.digits + string.punctuation + ' '
number = user_check()
password = random.sample(population=AVAILABLE,k=number)
print('computer password is {p}'.format(p=''.join(password)))
print()
time.sleep(3)
#Break a password
INDEX = 0
pass_list = [' ' for i in range(number)]
for i in range(number):
    for _,value in enumerate(AVAILABLE):
        pass_list[INDEX] = value
        time.sleep(0.03)
        print(pass_list)
        if pass_list[INDEX] == password[INDEX]:
            winsound.Beep(frequency=1000,duration=500)
            break
    INDEX += 1
print()
print('Your guess is {g}'.format(g=''.join(pass_list)))
print()
if (''.join(pass_list)) == (''.join(password)):
    print('Good job on breaking password')
else:
    print('They are not the same')    
```
