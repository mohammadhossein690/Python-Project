```python
# First we import random module
import random
```

```python
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':11, 'Queen':12, 'King':13, 'Ace':14}
sum_1 = sum_2 = 0
```

```python
#---------------------------- These are classes We defined ----------------------------
```

```python
class Card:
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank] 
    def __str__(self):
        return self.rank + ' of ' + self.suit
class Deck:
    def __init__(self):
        self.all_cards = [] 
        for suit in suits:
            for rank in ranks:
                self.all_cards.append(Card(suit,rank))             
    def shuffle(self):
        random.shuffle(self.all_cards)
    def deal_one(self):
        return self.all_cards.pop()  
```

```python
#---------------------------- These are functions We defined ----------------------------
```

```python
def are_you_ready():
    while True:
        ready = input('Are you ready(Y/N) : ').upper()
        if ready == 'Y':
            return True
        else:
            return False
def play_again():
    while True:
        play = input('You wanna play again(Y/N) : ').upper()
        if play == 'Y':
            return True
        else:
            return False
```

```python
# This is the main part of our game
print('Welcome to Black Jack game!!')
if are_you_ready():
    name_1 = input('Hey player one please Enter your name : ').title()
    name_2 = input('Hey player two please Enter your name : ').title()
    while True:
        while True:
            bet = input('How much do you wanna bet ? ')
            if bet.isdigit():
                bet = int(bet)
                break
            else:
                print('Please Enter number!!!')
        print('-'*30)
        my_deck = Deck()
        my_deck.shuffle()
        name_1_balnace = bet
        name_2_balnace = bet
        print(f'Hey {name_1} It\'s your turn!! ')
        Mylst1 = []
        while True: 
            choice1 = input('Do you wanna hit or stand : ').lower()
            print('-'*30)
            if choice1 == 'hit':
                a = str(my_deck.deal_one()).split()[0]
                if a == ranks[-1]:
                    if sum_1 <= 7:
                        values[a] = 14
                    else:
                        values[a] = 1
                Mylst1.append(values[a])
                print(Mylst1)
                sum_1 = sum(Mylst1)
                print(f'The sum is {sum_1}')
            else:
                print(f'The sum is {sum_1}')
                break
            if sum_1 > 21:
                print('BUSTT!!!!!\nYou loose the game')
                break
        print()
        print(f'Hey {name_2} It\'s your turn!! ')
        Mylst2 = []
        while True: 
            choice2 = input('Do you wanna hit or stand : ').lower()
            print('-'*30)
            if choice2 == 'hit':
                b = str(my_deck.deal_one()).split()[0]
                if b == ranks[-1]:
                    if sum_2 <= 7:
                        values[b] = 14
                    else:
                        values[b] = 1
                Mylst2.append(values[b])
                print(Mylst2)
                sum_2 = sum(Mylst2)
                print(f'The sum is {sum_2}')
            else:
                print(f'The sum is {sum_2}')
                break
            if sum_2 > 21:
                print('BUSTT!!!!!\nYou loose the game')
                break
        print()
        if sum_1 > sum_2 and sum_1 <= 21:
            name_2_balnace -= bet
            name_1_balnace += bet
            print(f'{name_1} win the game')
        elif sum_1 < sum_2 and sum_2 <= 21:
            name_1_balnace -= bet
            name_2_balnace += bet
            print(f'{name_2} win the game')
        elif sum_1 > 21 and sum_2 <= 21:
            name_1_balnace -= bet
            name_2_balnace += bet
            print(f'{name_2} win the game')
        elif sum_2 > 21 and sum_1 <= 21:
            name_2_balnace -= bet
            name_1_balnace += bet
            print(f'{name_1} win the game')  
        else:
            print('They are equal')
        print()
        print()
        print(f'{name_1}\'s balance is {name_1_balnace} $')
        print(f'{name_2}\'s balance is {name_2_balnace} $')
        if play_again() == False:
            print('Hope see you again!')
            break
else:
    print('Hope see you again!')
```
